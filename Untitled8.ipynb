```python
# Import required libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# --- Step 1: Create the dataset ---
data = {
    'Outlook': ['Sunny','Sunny','Overcast','Rain','Rain','Rain','Overcast','Sunny',
                'Sunny','Rain','Sunny','Overcast','Overcast','Rain'],
    'Temperature': ['Hot','Hot','Hot','Mild','Cool','Cool','Cool','Mild',
                    'Cool','Mild','Mild','Mild','Hot','Mild'],
    'Humidity': ['High','High','High','High','Normal','Normal','Normal','High',
                 'Normal','Normal','Normal','High','Normal','High'],
    'Windy': ['Weak','Strong','Weak','Weak','Weak','Strong','Strong','Weak',
              'Weak','Weak','Strong','Strong','Weak','Strong'],
    'Play': ['No','No','Yes','Yes','Yes','No','Yes','No',
             'Yes','Yes','Yes','Yes','Yes','No']
}

df = pd.DataFrame(data)
print("Dataset:\n", df, "\n")

# --- Step 2: Encode categorical variables ---
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])

# Split features and labels
X = df[['Outlook','Temperature','Humidity','Windy']]
y = df['Play']

# --- Step 3: Train-test split ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# --- Step 4: Train Naive Bayes model ---
model = GaussianNB()
model.fit(X_train, y_train)

# --- Step 5: Predict and evaluate ---
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')

print("Accuracy:", round(accuracy, 2))
print("F1 Score:", round(f1, 2))

# --- Step 6: Display confusion matrix ---
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['No','Yes'])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix - Weather Forecast (Play Tennis)")
plt.show()

# --- Step 7: Test on a single sample ---
sample = X_test.iloc[2].values.reshape(1, -1)
predicted = model.predict(sample)
print("\nActual Value:", y_test.iloc[2])
print("Predicted Value:", predicted[0])
```
