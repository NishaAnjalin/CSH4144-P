```python
m=np.array([[1,5,2],
            [4,7,4],
            [2,0,9]])
print(m)
```

```python
print('Matrix Transpose:\n',m.transpose(),'\n')
print('Matrix Determinant:\n',np.linalg.det(m),'\n') #performing linear algebra aperation
m_inv=np.linalg.inv(m)
print('Matrix Inverse:\n',m_inv,'\n')
iden_m=np.dot(m,m_inv)
iden_m=np.round(np.abs(iden_m),0)
print('Product of matrix and its inverse:\n',iden_m)
```

```python
eigen_vals, eigen_vecs=np.linalg.eig(m)
print('Eigen values:\n',eigen_vals,'\n')
print('Eigen vector:\n',eigen_vecs)
```

```python
U, S, VT =np.linalg.svd(m)
print('getting SVD output:-\n')
print('U:\n',U,'\n')
print('S:\n',S,'\n')
print('VT:\n',VT)
```

```python
#Descriptive lib
import scipy as sp
import numpy as np
```

```python
#get random data
nums=np.random.randint(1,20,size=(1,15))[0]
print('Data:',nums)
```

```python
#get descriptive stats
print('Mean:',sp.mean(nums))
print('Median:',sp.median(nums))
print('Mode:',sp.stats.mode(nums))
print('Standard Deviation:',sp.std(nums))
print('Variation:', sp.var(nums))
print('Skew:',sp.stats.skew(nums))
print('Kurtosis:',sp.stats.kurtosis(nums))
```

```python
print('Mean:',np.mean(nums))
print('Median:',np.median(nums))
print('Mode:',sp.stats.mode(nums))
print('Standard Deviation:',np.std(nums))
print('Variation:', np.var(nums))
print('Skew:',sp.stats.skew(nums))
print('Kurtosis:',sp.stats.kurtosis(nums))
```
